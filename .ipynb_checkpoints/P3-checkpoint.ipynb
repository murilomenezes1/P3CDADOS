**Correlação ITUB4 e Ibovespa**

```python
plt.scatter(Close_bovespa,Close_ITUB4,color="blue")
def reta(dados_x, dados_y):    
    a = dados_x.cov(dados_y) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear entre X3 e X4:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, alpha=0.8)
    plt.title('ITUB4 vs Ibovespa')
    plt.xlabel('Ibovespa')
    plt.ylabel('ITUB4')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color="red")

    plt.tight_layout()
    
reta(Close_bovespa,Close_ITUB4)
```

```python
print("Coeficiente de correlação: {}".format(Close_bovespa.corr(Close_ITUB4)))
```

A variação de preço da ação ITUB4 apresenta forte correlação linear positiva com a variação do indice bovespa, uma vez que o coeficiente de correlação linear entre as duas é igual a 0.8497349968894575

## Construindo o Modelo

**Importando as Bibliotecas Necessárias**

```python
import tensorflow as tf 
import numpy as np 


print(tf.__version__) # Para ter certeza que estamos rodando tensorflow 2.0 
```

**Determinando os Datasets de Treinamento e Validação do Modelo**

```python
Training_set = web.get_data_yahoo("ITUB4.SA",start="2001-01-01",end="2007-11-28")
Training_output = web.get_data_yahoo("ITUB4.SA",start="2008-01-01",end="2014-12-31")
```

```python
T_x = Training_set["Adj Close"]
T_y = Training_output["Adj Close"]

X_train = []
Y_train = []

for i in range(60,T_x.shape[0]):
    X_train.append(T_x[i-60:i])

X_train = np.array(X_train)
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1],1)

for i in range (60, T_y.shape[0]):
    Y_train.append(T_y[i])

Y_train = np.array(Y_train)

print("Shape de X_train: {}".format(X_train.shape))
print("Shape de Y_train: {}".format(Y_train.shape))
```

**O modelo**

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.LSTM(60, activation='relu',input_shape=(60,1), return_sequences=True))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.LSTM(60,activation='relu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(1))


adam = tf.keras.optimizers.Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999, amsgrad=False)


model.compile(optimizer=adam, loss="mean_squared_error")
model.summary()
history = model.fit(X_train,Y_train, epochs=100,validation_split=0.33)



```

**Diagnostico do modelo**

```python
plt.figure(figsize=(14,4))
plt.plot(history.history["loss"], label="Loss de Treinamento")
plt.plot(history.history["val_loss"], label = "Loss de Validação")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss do modelo")
plt.show()
```

**Previsão**

```python
X_val = web.get_data_yahoo("ITUB4.SA",start="2015-02-02",end="2016-12-26")
Y_val = web.get_data_yahoo("ITUB4.SA",start="2017-01-02",end="2018-11-14")
V_x = X_val["Adj Close"]
V_y = Y_val["Adj Close"]

X_pred = []
Y_pred = []


# Vetorizando o dataset
for i in range(60,V_x.shape[0]):
    X_pred.append(V_x[i-60:i])
    
X_pred = np.array(X_pred)
X_pred = X_pred.reshape(X_pred.shape[0],X_pred.shape[1],1)

for i in range(60, V_y.shape[0]):
    Y_pred.append(V_y[i])

Y_pred = np.array(Y_pred)


print(X_pred.shape)
print(Y_pred.shape)

```

```python
prediction = model.predict(X_pred)
prediction = prediction/1000
```

**Visualização da Previsão**

```python
plt.figure(figsize=(14,4))
plt.plot(prediction, label="Model Prediction",color='red')
# plt.grid(True)
# plt.show()
# plt.figure(figsize=(14,4))
plt.plot(Y_pred, label ="ITUB4 real stock price",color='blue')
plt.grid(True)
plt.show()

```

## Segunda Iteração do Modelo

**Normalizacão dos Datasets**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# NT_x = Training_set.drop(['High','Low','Open','Close','Volume'],axis=1)
# NT_y = Training_output.drop(['High','Low','Open','Close','Volume'],axis=1)

dataset = web.get_data_yahoo("ITUB4.SA",start="2001-01-01",end="2014-12-31").drop(['High','Low','Open','Close','Volume'],axis=1)

dataset = scaler.fit_transform(dataset)
dataset

# NT_x = scaler.fit_transform(NT_x)
# NT_y = scaler.fit_transform(NT_y)
```

**Vetorização dos datasets**

```python
X_train2 = []
Y_train2 = []

for i in range(60,dataset.shape[0]):
    X_train2.append(dataset[i-60:i])
    Y_train2.append(dataset[i,0])
    
    
X_train2 = np.array(X_train2)
Y_train2 = np.array(Y_train2)


print("Shape de X_train2: {}".format(X_train2.shape))
print("Shape de Y_train2: {}".format(Y_train2.shape))
```

**O modelo**

```python
model2 = tf.keras.Sequential()

model2.add(tf.keras.layers.LSTM(60,input_shape=(60,1), return_sequences=True))
model2.add(tf.keras.layers.Dropout(0.5))

model2.add(tf.keras.layers.LSTM(60,return_sequences=True))
model2.add(tf.keras.layers.Dropout(0.5))

model2.add(tf.keras.layers.LSTM(60))
model2.add(tf.keras.layers.Dropout(0.5))
model2.add(tf.keras.layers.Dense(1))

adam = tf.keras.optimizers.Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, amsgrad=False)


model2.compile(optimizer=adam, loss="mean_squared_error")
model2.summary()
history2 = model2.fit(X_train2,Y_train2, epochs=50,validation_split=0.33)
```

```python
plt.figure(figsize=(14,4))
plt.plot(history2.history["loss"], label="Loss de Treinamento")
plt.plot(history2.history["val_loss"], label = "Loss de Validação")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()


plt.title("Loss do modelo")
plt.show()
```
